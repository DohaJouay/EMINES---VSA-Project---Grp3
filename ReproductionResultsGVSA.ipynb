# Importing Librairies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Defining Variables and Data

```python
resultsvsa = pd.read_csv('vsaresults.txt', delimiter = '\t', header=None)
```

```python
percep1 = list([resultsvsa[0][0], resultsvsa[2][0],resultsvsa[4][0],resultsvsa[6][0]])
percep10 = list([resultsvsa[1][0], resultsvsa[3][0],resultsvsa[5][0],resultsvsa[7][0]])
linearhdc1 = list([resultsvsa[0][1], resultsvsa[2][1],resultsvsa[4][1],resultsvsa[6][1]])
linearhdc10 = list([resultsvsa[1][1], resultsvsa[3][1],resultsvsa[5][1],resultsvsa[7][1]])
rffhdc1 = list([resultsvsa[0][2], resultsvsa[2][2],resultsvsa[4][2],resultsvsa[6][2]])
rffhdc10 = list([resultsvsa[1][2], resultsvsa[3][2],resultsvsa[5][2],resultsvsa[7][2]])
rff1gvsa1 = list([resultsvsa[0][3], resultsvsa[2][3],resultsvsa[4][3],resultsvsa[6][3]])
rff1gvsa10 = list([resultsvsa[1][3], resultsvsa[3][3],resultsvsa[5][3],resultsvsa[7][3]])
rff2gvsa1 = list([resultsvsa[0][4], resultsvsa[2][4],resultsvsa[4][4],resultsvsa[6][4]])
rff2gvsa10 = list([resultsvsa[1][4], resultsvsa[3][4],resultsvsa[5][4],resultsvsa[7][4]])
rff3gvsa1 = list([resultsvsa[0][5], resultsvsa[2][5],resultsvsa[4][5],resultsvsa[6][5]])
rff3gvsa10 = list([resultsvsa[1][5], resultsvsa[3][5],resultsvsa[5][5],resultsvsa[7][5]])
rff4gvsa1 = list([resultsvsa[0][6], resultsvsa[2][6],resultsvsa[4][6],resultsvsa[6][6]])
rff4gvsa10 = list([resultsvsa[1][6], resultsvsa[3][6],resultsvsa[5][6],resultsvsa[7][6]])
rff5gvsa1 = list([resultsvsa[0][7], resultsvsa[2][7],resultsvsa[4][7],resultsvsa[6][7]])
rff5gvsa10 = list([resultsvsa[1][7], resultsvsa[3][7],resultsvsa[5][7],resultsvsa[7][7]])
```

```python
mnistgvsa1 = list([resultsvsa[4][3], resultsvsa[4][4],resultsvsa[4][5],resultsvsa[4][6],resultsvsa[4][7]])
mnistgvsa10 = list([resultsvsa[5][3], resultsvsa[5][4],resultsvsa[5][5],resultsvsa[5][6],resultsvsa[5][7]])
fmnistgvsa1 = list([resultsvsa[6][3], resultsvsa[6][4],resultsvsa[6][5],resultsvsa[6][6],resultsvsa[6][7]])
fmnistgvsa10 = list([resultsvsa[7][3], resultsvsa[7][4],resultsvsa[7][5],resultsvsa[7][6],resultsvsa[7][7]])
isoletgvsa1 = list([resultsvsa[0][3], resultsvsa[0][4],resultsvsa[0][5],resultsvsa[0][6],resultsvsa[0][7]])
isoletgvsa10 = list([resultsvsa[1][3], resultsvsa[1][4],resultsvsa[1][5],resultsvsa[1][6],resultsvsa[1][7]])
ucihargvsa1 = list([resultsvsa[2][3], resultsvsa[2][4],resultsvsa[2][5],resultsvsa[2][6],resultsvsa[2][7]])
ucihargvsa10 = list([resultsvsa[3][3], resultsvsa[3][4],resultsvsa[3][5],resultsvsa[3][6],resultsvsa[3][7]])
```

```python
CDC = pd.read_csv('CDC.txt', delimiter = '\t', header=None)
CDCmnist = CDC[2]
CDCfmnist = CDC[3]
CDCisolet = CDC[0]
CDCucihar = CDC[1]
```

# MNIST DataSet

```python
# Data for the plot
group_order = [2**i for i in range(1, 6)]  # [2^1, 2^2, 2^3, 2^4, 2^5]
x_positions = np.arange(len(group_order))
one_epoch_accuracy = mnistgvsa1
final_accuracy = mnistgvsa10
cdc_values = CDCmnist

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot the accuracies (primary Y-axis)
ax1.plot(x_positions, one_epoch_accuracy, 'o-', color='red', label='1-Epoch Accuracy')
ax1.plot(x_positions, final_accuracy, 'v--', color='blue', label='Final Accuracy')
ax1.set_xlabel('Group Order')
ax1.set_ylabel('Test Accuracy (%)')
ax1.set_xticks(x_positions)
ax1.set_xticklabels(group_order)
ax1.legend(loc='upper left')

# Create a secondary Y-axis for CDC values
ax2 = ax1.twinx()
cdc_bar = ax2.bar(x_positions, cdc_values, color='lightgreen', alpha=0.5, label='CDC')
ax2.set_ylabel('Circuit-Depth Complexity', color='red')
ax2.tick_params(axis='y', colors='red')

# Add a legend for the bar
lines_labels = ax1.get_legend_handles_labels()
bars_labels = [(cdc_bar[0], 'CDC')]
ax1.legend(lines_labels[0] + [bar[0] for bar in bars_labels], lines_labels[1] + [bar[1] for bar in bars_labels], loc='upper left')

fig.tight_layout()
plt.title('Cyclic Group VSA on MNIST')

# Show the plot
plt.grid(True)
plt.show()
```
